```python
import math 
import numpy as np
import matplotlib.pyplot as plt

def deriv(f, x0, h) -> float:
    """
        This function takes in a function as well as the x not and a variety h step size 
        values to calculate an approximated derivative value. Given that data it returns 
        the derivative. 
    """  
    return (f(x0+h) - f(x0))/h

def derivTS(f, x0, h) -> float:
    """
        This function is similar to the deriv function but differs in 
        calculating an approximated derivative value using the Taylor seires. 
    """
    return (f(x0+h) - f(x0-h))/(2*h)

def errorOfNumericDeriv(d_analytical, d_numerical):
    """
        This function is used to calculate the error in approximating the derivative value 
        versus the actual derivative value.
    """
    return math.fabs((d_analytical - d_numerical) / d_analytical)

# Variable constants 
x0 = 0.1
d_analytical = math.cos(x0)
h = [0.1, 0.02,0.003, 0.24, 0.805]
h.sort()

# Arrays for storing error of numerical derivaitve for each function
arrayTs = []
arrayDeriv = []

for elem in h: 
    derivOut = deriv(math.sin, x0, elem)
    derivTSOut = derivTS(math.sin, x0 ,elem)
    
    errorNumDer = errorOfNumericDeriv(d_analytical, derivOut)
    errorNumDerTS = errorOfNumericDeriv(d_analytical, derivTSOut)
    
    arrayTs.append(errorNumDer)
    arrayDeriv.append(errorNumDerTS)

# Creating graph of data
plt.plot(h, arrayTs)
plt.xlabel('h step values')
plt.plot(h, arrayDeriv)
plt.ylabel('error of approximate derivative')
plt.show()

    
 
        
```
